```python
import pandas as pd
import matplotlib.pyplot as plt

# Crear un DataFrame con los datos proporcionados
datos = {
    "Nombre": ["Ana", "Luis", "Paula"],
    "Apellido": ["García", "Martínez", "López"],
    "Edad": [20, 21, 22],
    "Nota Media": [8.5, 7.0, 9.0],
    "Asignaturas Aprobadas": [5, 4, 5]
}

df = pd.DataFrame(datos)

# Gráfica 1: Distribución de notas medias
plt.figure(figsize=(8, 5))
plt.bar(df["Nombre"], df["Nota Media"], color="skyblue", edgecolor="black")
plt.title("Distribución de Notas Medias", fontsize=14)
plt.xlabel("Alumno", fontsize=12)
plt.ylabel("Nota Media", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()

# Gráfica 2: Asignaturas aprobadas por alumno
plt.figure(figsize=(8, 5))
plt.bar(df["Nombre"], df["Asignaturas Aprobadas"], color="lightgreen", edgecolor="black")
plt.title("Asignaturas Aprobadas por Alumno", fontsize=14)
plt.xlabel("Alumno", fontsize=12)
plt.ylabel("Asignaturas Aprobadas", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()

# Gráfica 3: Porcentaje de aprobados y suspensos (circular)
resultados = ["Aprobados", "Suspensos"]
cantidad = [len(df[df["Nota Media"] >= 5]), len(df[df["Nota Media"] < 5])]
plt.figure(figsize=(7, 7))
plt.pie(cantidad, labels=resultados, autopct="%1.1f%%", colors=["lightblue", "salmon"], startangle=90)
plt.title("Porcentaje de Aprobados y Suspensos", fontsize=14)
plt.tight_layout()
plt.show()

# Gráfica 4: Notas más altas y más bajas
plt.figure(figsize=(8, 5))
plt.bar(["Nota más alta", "Nota más baja"], [df["Nota Media"].max(), df["Nota Media"].min()],
        color=["gold", "tomato"], edgecolor="black")
plt.title("Notas Extremas (Alta vs Baja)", fontsize=14)
plt.ylabel("Nota", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```
