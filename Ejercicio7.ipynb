```python
import pandas as pd
from matplotlib import pyplot as plt
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas

# **Leer el archivo CSV (Datos de los estudiantes)**
df = pd.read_csv('ruta/a/tu/archivo.csv')  # Asegúrate de usar la ruta correcta

# **Ejercicio 5: Análisis de los datos**
# Calcular la media de la nota por cada alumno
df['Mitjana'] = df['Nota Media']

# Calcular el porcentaje de aprobados
aprovats = (df['Mitjana'] >= 5).mean() * 100

# Nota más alta y más baja
nota_max = df['Mitjana'].max()
nota_min = df['Mitjana'].min()

# Mostrar resultados
print(f"Mitjana global: {df['Mitjana'].mean():.2f}")
print(f"Percentatge d'aprovats: {aprovats:.2f}%")
print(f"Nota més alta: {nota_max}, Nota més baixa: {nota_min}")

# **Ejercicio 6: Generación de Gráficos**
# Gráfica de barras para las notas de cada alumno
df.set_index('Nombre')['Nota Media'].plot(kind='bar', figsize=(10, 6))
plt.title('Notas de los Estudiantes')
plt.ylabel('Nota Media')
plt.tight_layout()
plt.savefig('grafico_barras.png')
plt.close()

# Gráfica de líneas de la evolución de las notas
df.set_index('Nombre')['Nota Media'].plot(kind='line', marker='o', figsize=(10, 6))
plt.title('Evolución de las Notas de los Estudiantes')
plt.ylabel('Nota Media')
plt.tight_layout()
plt.savefig('grafico_lineas.png')
plt.close()

# Gráfica circular para la distribución de las notas
df['Nota Media'].plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8), startangle=90)
plt.title('Distribución de las Notas Medias')
plt.ylabel('')
plt.tight_layout()
plt.savefig('grafico_pastel.png')
plt.close()

# **Ejercicio 7: Informe en PDF**
# Crear el PDF
c = canvas.Canvas("informe_final.pdf", pagesize=letter)

# Título
c.drawString(100, 750, "Informe Final - Ejercicios 5, 6 y 7")

# Resultados del Ejercicio 5
c.drawString(100, 700, "Resultados del Ejercicio 5:")
text = c.beginText(100, 680)
text.setFont("Helvetica", 10)
text.textLines(
    f"Mitjana global: {df['Mitjana'].mean():.2f}\n"
    f"Percentatge d'aprovats: {aprovats:.2f}%\n"
    f"Nota més alta: {nota_max}, Nota més baixa: {nota_min}\n"
)
c.drawText(text)

# Gráficos del Ejercicio 6
c.drawString(100, 620, "Gráficos del Ejercicio 6:")
c.drawImage("grafico_barras.png", 100, 400, width=300, height=200)
c.drawImage("grafico_lineas.png", 100, 180, width=300, height=200)

# Guardar el PDF
c.save()
```
